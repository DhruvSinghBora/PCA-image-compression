```python
# image compression using PCA 
#pca = principle component analysis
# unsupervised machine learning algorithm used for dimensionality reduction
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
#reading image
img = cv2.cvtColor(cv2.imread('image3.jpeg'), cv2.COLOR_BGR2RGB)
```

```python
#(resolution of image, number of channels(rgb))
img.shape
```

```python
#we will reduce the dimensionality of image
# we will treet column as dataset
plt.imshow(img)
```

```python
# split rgb channels
r, g, b = cv2.split(img)
```

```python
r, g, b = r/255, g/255, b/255
```

```python
plt.imshow(r)
```

```python
plt.imshow(g)
```

```python
plt.imshow(b)
```

```python
pca_components = 30

pca_r = PCA(n_components=pca_components)
reduced_r = pca_r.fit_transform(r)
pca_g = PCA(n_components=pca_components)
reduced_g = pca_g.fit_transform(g)
pca_b = PCA(n_components=pca_components)
reduced_b = pca_b.fit_transform(b)
```

```python
reduced_r.shape
```

```python
combined = np.array([reduced_r, reduced_g, reduced_b])
```

```python
reconstructed_r = pca_r.inverse_transform(reduced_r)
reconstructed_g = pca_g.inverse_transform(reduced_g)
reconstructed_b = pca_b.inverse_transform(reduced_b)
```

```python
reconstructed_r.shape
```

```python
plt.imshow(reconstructed_r)
```

```python
img_reconstructed = (cv2.merge((reconstructed_r, reconstructed_g, reconstructed_b)))
```

```python
plt.imshow(img_reconstructed)
```
